```python
from collections import deque
def waterJugProblem(c1,c2,goal):
    queue = deque() # 0,0
    visited=set()
    actions=[]
    queue.append((0,0)) #X=0, Y=0
    visited.add((0,0))

    while queue:
        jug1, jug2 = queue.popleft()
        actions.append((jug1,jug2))
        if jug1 == goal or jug2 == goal:
            print("Solution Found")
            for action in actions:
                print(action)
            return True
        # Rules
        rules=[
            (c1, jug2), #1. Fill X
            (jug1, c2), #2. Fill Y
            (0, jug2), #3. Empty X
            (jug1, 0), #4. Empty Y
            (0, jug1+jug2), #5. Pour from X to Y
            (jug1+jug2, 0), #6. Pour from Y to X
            (abs(jug1-(c2-jug2)),c2, ),#7. Pour from X to Y untill Y is full
            (c1,abs(jug2-(c1-jug1))), #8. Pour from Y to X untill X is full
        ]
        for state in rules:
            if state not in visited:
                visited.add(state)
                queue.append(state)
    print("No Solution Found")
    return False
                
jug1Capacity1=4
jug2Capacity2=3
target=1

waterJugProblem(jug1Capacity1, jug2Capacity2, target)    
```
